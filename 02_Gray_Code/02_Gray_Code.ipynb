<h3>Gray Code</h3>

A Gray code is a list of all 2^n bit strings of length n, where any two successive strings differ in exactly one bit (i.e., their Hamming distance is one).

Your task is to create a Gray code for a given length n.

<h5>Input</h5>
The only input line has an integer n.

<h5>Output</h5>
Print 2^n lines that describe the Gray code. You can print any valid solution.

<p></p>
<a href="https://cses.fi/problemset/task/2205"> Problem </a>
<a href="https://cses.fi/paste/3fb99c4d72e34279ecb62b/"> Solution </a>
<p></p>

```python
%%bash
cat << 'EOF' > gray-code.cpp

#include <bits/stdc++.h>
#include <string>
#define ll              long long
#define mfor(i, n)      for(ll i=1; i<=n; ++i)
#define nfor(i, n)      for(ll i=n; i>=1; --i)
#define ffor(i, m, n)   for(ll i=m; i<=n; ++i)
#define bfor(i, m, n)   for(ll i=m; i>=n; --i)
using namespace std;
 
const ll maxn = (1<<16) + 9;
 
string gray[maxn];
 
int main(){
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);
    cout.tie(NULL);
 
    ll n;
    cin >> n;
 
    ll i = 1;
    while (i < (1<<n)){
        mfor(j, i)
            gray[i+j] = "1" + gray[i-j+1];
        mfor(j, i)
            gray[j] = "0" + gray[j];
        i <<= 1;
    }
 
    mfor(i, 1<<n)
        cout << gray[i] << endl;
 
    return 0;
}
EOF

cat << 'EOF' > input.txt
3
EOF

g++ gray-code.cpp -o gray-code
./gray-code < input.txt
```

```python
%%bash
cat << 'EOF' > gray-code-zare.c

#include "stdio.h"
#define ll              long long
#define inp(n)          scanf("%lli", &n)
#define out(n)          printf("%lli", n)
#define endl            printf("\n")
#define mfor(i, n)      for(ll i=0; i<n; ++i)
#define maxn            (2*100*1000)


int main(){
    ll n;
    inp(n);

    mfor(i, 1<<n){
        mfor(j, n)
            out(((i ^ (i>>1)) & (1<<(n-j-1))) != 0);
        endl;
    }
}
EOF

cat << 'EOF' > input.txt
3
EOF

g++ gray-code-zare.c -o gray-code-zare
./gray-code-zare < input.txt
```

<h3>Sum of Two Values</h3>
You are given an array of n integers, and your task is to find two values (at distinct positions) whose sum is x.

<h5>Input</h5>
The first input line has two integers n and x: the array size and the target sum.
The second line has n integers a_1,a_2,\dots,a_n: the array values.

<h5>Output</h5>
Print two integers: the positions of the values. If there are several solutions, you may print any of them. If there are no solutions, print IMPOSSIBLE.
