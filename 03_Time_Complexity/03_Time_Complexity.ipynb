<h3>Sum of Two Values</h3>
You are given an array of n integers, and your task is to find two values (at distinct positions) whose sum is x.

<h5>Input</h5>
The first input line has two integers n and x: the array size and the target sum.
The second line has n integers a_1,a_2,\dots,a_n: the array values.

<h5>Output</h5>
Print two integers: the positions of the values. If there are several solutions, you may print any of them. If there are no solutions, print IMPOSSIBLE.

```python
#include "stdio.h"
#include <stdlib.h>
#define ll              long long
#define inp(n)          scanf("%lli", &n)
#define out(n)          printf("%lli ", n)
#define endl            printf("\n")
#define mfor(i, n)      for(ll i=0; i<n; ++i)
#define maxn            (2*100*1000)
 
 
 
struct pair_ll{
    ll left;
    ll right;
};
 
// int comp (const void * elem1, const void * elem2) {
//     ll f = *((ll*)elem1);
//     ll s = *((ll*)elem2);
//     if (f > s) return  1;
//     if (f < s) return -1;
//     return 0;
// }
 
int comp (const void * elem1, const void * elem2) {
    ll f = (*((struct pair_ll*)elem1)).left;
    ll s = (*((struct pair_ll*)elem2)).left;
    if (f > s) return  1;
    if (f < s) return -1;
    return 0;
}
 
 
struct pair_ll arr[maxn];
 
int main(){
    ll n, x;
    inp(n); inp(x);
    
    mfor(i, n){
        inp(arr[i].left);
        arr[i].right = i;
    }
 
    qsort(arr, n, sizeof(*arr), comp);
    // mfor(i, n){
    //     out(arr[i].left);
    //     out(arr[i].right);
    //     endl;
    // }
 
    ll p1 = 0, p2 = n-1, t;
    while (p1 < p2){
        // out(p1); out(p2); endl;
        // inp(t);
        ll sum = arr[p1].left + arr[p2].left;
        if (sum < x) p1++;
        else if (sum > x) p2--;
        else {
            out(arr[p1].right+1);
            out(arr[p2].right+1);
            endl;
            break;
        }
    }
    if (p1 >= p2) printf("IMPOSSIBLE\n");
}
```
